```python
import numpy as np
import pandas as pd
from glob import glob
```

```python
in_dir = '../Pipeline/notes-time/individual/'
out_dir = '../Pipeline/pitch-class-time/individual/'
```

```python
########################################
# get interval time series
# intervals for successive notes
########################################

def get_successive_intervals(filepath):
    # read in input
    notes = pd.read_csv(filepath)
    
    # construct pitch-class dataframe
    notes_long = pd.melt(notes,notes.columns[0],notes.columns[1:],'note','velocity')
    notes_long.note = [int(i)%12 for i in notes_long.note]
    
    # write dataframe to csv
    outname = filepath.split('/')[-1].replace('-notes.csv','-pitch-class.csv')
    notes_long.to_csv(out_dir+outname,index=False)
    return      
        
```

```python
# main
infiles = glob(in_dir+"*")
i = 0
for f in infiles:
    print(str(i)+"/"+str(len(infiles)))
    print(f)
    get_successive_intervals(f)
    i += 1
```
