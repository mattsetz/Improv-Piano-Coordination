```python

# A) {C,C,C,C,C,C,C,C,C,C} compared to {C,C#,D,D#,E,F,F#,G,G#,A,A#,B}  - should be high distance
# B) {C,C,C,C,C,C,C,C,C,C} compared to {C,C,C,C,C,C,C,C,C,C}  - very low distance
# C) {C,C#,D,D#,E,F,F#,G,G#,A,A#,B} compared to {C,C#,D,D#,E,F,F#,G,G#,A,A#,B} - medium low distance
# D) {C,C,C,C,C,C,C,C,C,C} compared to {F#,F#,F#,F#,F#,F#,F#,F#,F#,F#,F#,F#} - high distance
# E) {C,C,C,C,C,C,C,C,C,C} compared to {G,G,G,G,G,G,G,G,G,G,G,G,G} - low distance
# F) {C,E,G,C,E,G} compared to {C,E,G,C,E,G} - very low distance
# G) {C,E,G,C,E,G} compared to {G,B,D,G,B,D} - medium low distance
# H) {C,E,G,C,E,G} compared to {F#,A#,C#,F#,A#,C#} - high distance
```

```python
# Test Cases
c = [10]+[0]*11 # 10 Cs
serial = [1]*12 # 12-tone row
fsharp = [0]*6+[10]+[0]*5 # 10 f sharps
g = [0]*7+[10]+[0]*4
cmaj = [2,0,0,0,2,0,0,2,0,0,0,0]
cmin = [2,0,0,2,0,0,0,2,0,0,0,0]
gmaj = [0,0,2,0,0,0,0,2,0,0,0,2]
fsharpmaj = [0,2,0,0,0,0,2,0,0,0,2,0]
```

```python
func_ids = ["NORM_MIN","SIMPLE","NOVELTY","WEIGHTED_NORM"]

test_cases = {"c-serial":(c,serial),
              "c-c":(c,c),
              "serial-serial":(serial,serial),
              "c-fsharp": (c,fsharp),
              "c-g": (c,g),
              "cmaj-cmaj": (cmaj,cmaj),
              "cmaj-gmaj": (cmaj,gmaj),
              "cmaj-fsharpmaj": (cmaj,fsharpmaj),
              "cmaj-cmin": (cmaj,cmin)}
```

```python
# look at distances assigned different functions
for func in func_ids:
    print(func)
    distances = []
    for t in test_cases:
        distances.append(inter_cloud_distance(func,test_cases[t][0],test_cases[t][1],diss_dict_chew_2_15))
    print(distances)
    plt.figure(figsize=[20,5])
    plt.bar(range(len(distances)),distances,tick_label=list(test_cases.keys()))
    plt.show()
```
